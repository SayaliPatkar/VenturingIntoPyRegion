```python
# Adding iterator to your own class and make it iterable
class Sentence:
    def __init__(self, sentence):
        self.sentence = sentence.split()
        self.start = 0
        self.end = len(self.sentence)
    def __iter__(self):
        return self
    
    def __next__(self):
        if self.start >= self.end:
            raise StopIteration
        current = self.sentence[self.start]
        self.start += 1
        return current
        
    
sentence_instance = Sentence("Love looks not with the eyes, but with the mind; and therefore is winged Cupid painted blind.")
for word in sentence_instance:
    print(word)
```

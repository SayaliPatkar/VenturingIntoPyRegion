```python
# Classes are blueprints describing proprties (attributes) and behaviours (methods) of an object
class Student:
    # class variables
    num_students = 0 # to be raised for every new instance as soon as instance is created
    fee_raise = 1.02 # default 2% raise in fees every year
    
    # constructin dunder method __init__
    def __init__ (self, first, last, grade, fees):
        self.first = first
        self.last = last
        self.grade = grade
        self.fees = fees
        self.email = first + '.' + last + '@university.com'
        
        # here we do not use self.num_students as 
        # we need this variable to increment for every new instance and should be same for all instances at any given time
        Student.num_students += 1
        
    # every class method has minimum one argument self
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    def revised_fees(self):
        return self.fees * self.fee_raise
    
class SchoolStudent(Student):
    # overrides fee_raise for all instances of SchoolStudent
    # does not affect fee_raise for parent class Student
    fee_raise = 1.05
    

s1 = SchoolStudent('Sayali','Patkar','1.7', 200)
print(s1.email)

# help function to see all own and inheritance imparted properties 
print(help(SchoolStudent))

print(help(Student))
```

```python
# Classes are blueprints describing proprties (attributes) and behaviours (methods) of an object
class Student:
    # class variables
    num_students = 0 # to be raised for every new instance as soon as instance is created
    fee_raise = 1.02 # default 2% raise in fees every year
    
    # constructin dunder method __init__
    def __init__ (self, first, last, grade, fees):
        self.first = first
        self.last = last
        self.grade = grade
        self.fees = fees
        self.email = first + '.' + last + '@university.com'
        
        # here we do not use self.num_students as 
        # we need this variable to increment for every new instance and should be same for all instances at any given time
        Student.num_students += 1
        
    # every class method has minimum one argument self
    def fullname(self):
        return '{} {}'.format(self.first, self.last)
    
    def revised_fees(self):
        return self.fees * self.fee_raise
    
class SchoolStudent(Student):
    # overrides fee_raise for all instances of SchoolStudent
    # does not affect fee_raise for parent class Student
    fee_raise = 1.05
    
    def __init__ (self, first, last, grade, fees, school):
        super().__init__(first, last, grade, fees)
        self.school = school
        
class KitaStudent(Student):
    pass

ss1 = SchoolStudent('Neena','Bansal','2.0', 100, 'our lady of salvation')
print(ss1.email)
print(ss1.school)

ks1 = KitaStudent('Geeta','Ghosh','1.0', 500 )

# help function to see all own and inheritance imparted properties 
print(help(SchoolStudent))
```

```python
#isinstance() and issubclass()
# is ks1 instance of KitaStudent
print(isinstance(ks1,KitaStudent))

print(isinstance(ks1,Student))

print(isinstance(ks1,SchoolStudent))

# is KitaStudent subclass of Student
print(issubclass(KitaStudent,Student))

print(issubclass(SchoolStudent,Student))

print(issubclass(KitaStudent,SchoolStudent))
```
